## 9.4- Use el metodo grafico para resolver el siguiente sistema:
\begin{align*}
     4x_1 - 8x_2 = -24\\
     x_1 + 6x_2 = 34
\end{align*}

```python
import math
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline
def grafica():
    x=np.arange(3,6,0.5)
    y=2*x-6
    z=(34-6*x)
    ptl.plot(x,y,'crimson',label='X1')
    ptl.plot(x,z,'SpringGreen',label='X2')
    ptl.title("Ejercicio 9.4", fontsize = 20)
    ptl.legend(loc="upper right")
    ptl.grid(True)
    ptl.show()
grafica()
```

## Compruebe el resultado por medio de sustituirlo en las ecuaciones.
#### Respuesta:
\begin{align*}
    x_1= 4\\
    x_2= 5
\end{align*}
#### *Remplazando en la ecuación (1)
\begin{align*}
    4(4) - 8(5)=-24\\
    16 - 40=-24
\end{align*}
#### *Remplazando en la ecuación (2)
\begin{align*}
    4 + 6(5)=34\\
    4 + 30=34
\end{align*}
