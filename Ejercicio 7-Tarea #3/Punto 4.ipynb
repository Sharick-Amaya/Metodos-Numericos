# 5.1- Determine las raíces reales de f(x) = –0.5x^2 + 2.5x + 4.5

```python
import math as mt
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
a=-0.5
b=2.5
c=4.5
x1=(-b+mt.sqrt(b**(2)-(4*a*c)))/2*a
x2=-2*c/(b-mt.sqrt(b**(2)-(4*a*c)))

xl=5.0
xu=10.0
xr=0.0
fxl=0.0
fxu=0.0
vver=x2#LA RAIZ Verdadera
ter=1
erpa=0.0

print ("Las raices son: x1= %f"%(x1))
print ("                x2= %f"%(x2))
print()
print("-Aproximaremos x2 por bisección a 3 términos")
print ("+---------------------------------------------------------------------------------------+")
print ("|Término|      Xl      |       Xu       |     Xr     |        εa        |       εt      |")
print ("+---------------------------------------------------------------------------------------+")

while(ter<4):
    val_ante=xr
    fxl=(a*xl**(2))+(b*xl)+c
    fxu=(a*xu**(2))+(b*xu)+c
    xr=(xl+xu)/2
    fxr=(a*xr**(2))+(b*xr)+c
    ev=vver-xr
    erpv=abs(ev/vver)*100
    erpa=abs((xr-val_ante)/(xr))*100
    
    print ("|   %d \t|   %f   |   %f\t|  %f  |    %f\t|     %1.3f\t|" %(ter, xl, xu, xr, erpa, erpv))
    print ("+---------------------------------------------------------------------------------------+")
    if(fxl*fxr<0):
        xu=xr
    if(fxl*fxr>0):
        xl=xr
    if (erpa<es or fxl*fxr==0 ):
        break
    ter=ter+1
print ("Valor Aproximado=  %1.3f"%(xr))
```

# 5.13- La velocidad v de un paracaidista que cae está dada por V=((gm/c)(1-e^-(c/m)t)) donde g = 9.8 m/s^2. Para un paracaidista con coeficiente de arrastre de c = 15 kg/s, calcule la masa m de modo que la velocidad sea v = 35 m/s en t =9s. Utilice el método de la falsa posición para determinar m a un nivel de εs = 0.1%.

```python
import math as mt

es=0.1
g=9.8
v=35.0
c=15.0
t=9.0
ml=50
mu=70
mr=0.0
fxl=0.0
fxu=0.0
ter=1
erpa=0.0

print ("+-----------------------------------------------------------------------+")
print ("|Término|      ml      |       mu       |     mr     |        εa        |")
print ("+-----------------------------------------------------------------------+")

while(True):
    val_ante=mr
    fxl=((g*ml/c)*(1-(mt.e**(-(c*t)/ml))))-v
    fxu=((g*mu/c)*(1-(mt.e**(-(c*t)/mu))))-v
    mr=mu-(fxu*(ml-mu)/(fxl-fxu))
    fxr=((g*mr/c)*(1-(mt.e**(-(c*t)/mr))))-v
    erpa=abs((mr-val_ante)/(mr))*100
   
    print ("|   %d \t|   %f   |   %f\t|  %f  |    %f\t|" %(ter, ml, mu, mr, erpa))
    print ("+-----------------------------------------------------------------------+")
    
    if(fxl*fxr<0):
        mu=mr
    if(fxl*fxr>0):
        ml=mr
    if (fxl*fxr==0 or erpa<es):
        break
    ter=ter+1
print ("Valor de m Aproximado=  %1.3f"%(mr))
```
