```python
import matplotlib.pyplot as plt

g=9.8
m=68.1
c=12.5
e=2.7

t1= 0
v1= 0
t2= 2
v2= 0

vel1= []
tie1= []
vel2= [] 
tie2= []

print ("|\tt\t|\tV(t)A\t\t|\tv(t)M.N\t\t|")
print ("-----------------------------------------------------------------")
while t1<=110:
    v1=(g*m/c)*(1-e**(-(c/m)*t1))
    if t1%2==0:
        vel2.append(v2)
        print ("|\t%d\t|\t%f\t|\t%f\t|" %(t1,v1,v2))
        v2=v2+(g-(c/m*v2))*(t2-(t2-2))
        tie2.append(t1)
        t2+=2
    else:
        print ("|\t%d\t|\t%f\t|\t---------       |" %(t1,v1))
    vel1.append(v1)
    tie1.append(t1)
    t1+=1
plt.plot(tie2, vel2, 'r.', label='Velocidad Metodos Numericos')
plt.plot(tie1, vel1, 'k-', label='Velocidad Analitica')
plt.title("VELOCIDAD DEL PARACAIDISTA")
plt.legend(loc=2)
plt.xlabel("Tiempo (s)")
plt.xlim(0, 110)
plt.ylabel("Velocidad (m)")
plt.ylim(0, 70)
plt.show()
```
