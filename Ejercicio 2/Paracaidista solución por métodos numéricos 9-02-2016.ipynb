```python
import matplotlib.pyplot as plt

g=9.8
m=68.1
c=12.5
e=2.7
t=0

ta = 0
va = 0
tmn = 2
vmn = 0

vela = []
tiea = []
velmn = [] 
tiemn= []

print ("t\t|\tv(t)a\t\t|\tv(t)mn")
print ("-------------------------------------------------")
while ta<=102:
    va=(g*m/c)*(1-e**(-(c/m)*ta))
    if ta%2==0:
        velmn.append(vmn)
        print ("%d\t|\t%f\t|\t%f" %(ta,va,vmn))
        vmn=vmn+(g-(c/m*vmn))*(tmn-(tmn-2))
        tiemn.append(ta)
        tmn+=2
    else:
        print ("%d\t|\t%f\t|\t--.---" %(ta,va))
    vela.append(va)
    tiea.append(ta)
    ta+=1
plt.plot(tiemn, velmn, 'bo', label='Velocidad Met. Num.')
plt.plot(tiea, vela, 'k', label='Velocidad Analitica')
plt.title("Velocidad del paracaidista")
plt.legend(loc=2)
plt.xlabel("Tiempo (s)")
plt.xlim(0, 110)
plt.ylabel("Velocidad (m)")
plt.ylim(0, 70)
plt.show()
```
