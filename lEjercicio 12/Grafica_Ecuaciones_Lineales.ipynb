```python
import math
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline

def graf1():
    ptl.title("Solucion_Grafica_Ecuaciones_Lineales")    
    x=np.arange(-2,6,0.1)
    y=((-3*x)/2)+9
    z=(x/2)+1
    ptl.plot(x,y,'-b')
    ptl.plot(x,z,'-r')
    ptl.plot(4,3,'o')
    ptl.grid(True)
    ptl.show()
    
def graf2():
    ptl.title("Sistema_Sin_Solucion")
    x=np.arange(-2,6,0.1)
    y=(x/2)+1
    z=(1/2)+(x/2)
    ptl.plot(x,y,'-b')
    ptl.plot(x,z,'-r')
    ptl.grid(True)
    ptl.show()
    
    ptl.title("Sistema_Con_Infinitas_Soluciones")
    x=np.arange(-2,6,0.1)
    w=1+(x/2)
    q=1+(x/2)
    ptl.plot(x,w,'-b')
    ptl.plot(x,q,'-r')
    ptl.grid(True)
    ptl.show()
    
    ptl.title("Sistema_Sin_Solucion_Visible_Graficamente")
    x=np.arange(-2,6,0.1)
    r=(1.1)+(2.3*x/5)
    t=(1)+(x/2)
    ptl.plot(x,r,'-b')
    ptl.plot(x,t,'-r')
    ptl.grid(True)
    ptl.show()
   
    
    
graf1()
graf2()
```
