```python
import math
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline

def graf1():
    ptl.title("Ecuaciones Algebraicas Lineales")    
    x=np.arange(-2,6,0.1)
    y=((-3*x)/2)+9
    z=(x/2)+1
    ptl.plot(x,y,'-g')
    ptl.plot(x,z,'-b')
    ptl.plot(4,3,'or')
    ptl.grid(True)
    ptl.show()

graf1()
```
