```python
import math

m=68.1 #Masa del paracaidista
c=12.5 #Coheficiente de resistencia
g=9.8 #Gravedad
t=0.0 #Se calcula desde el tiempo 0
i=1 #Intervalos
v2=0#Velocidad aproximada
v2ant=0#Velocidad aproximada anterior
ev= 0.0   #Error verdadero
erpv= 0.0 #Error relativo porcentual verdadero 
ea= 0.0 #Error relativo porcentual apróximado 
es=0.00005 #Error tolerado
cont=0

print ("-----------------------------------------------------------------------------------------")
print "|   %-3s\t|     %-10s|    %-9s  |      %-9s|      %-10s|      %-10s|"%('t','V(t)A','V(t)M.N','Et','εt','εa') 
print ("-----------------------------------------------------------------------------------------")

while (ea>es or ea==0):
    aux=(c/m*t)
    v1=((m*g)/(c))*(1-(math.exp(-1*aux)))
    if(t==0):
        v2=0.0
    else:
        v2=v2+(g-(c/m)*v2)*((t+i)-t)     
        
    if(v2==0):
        ea=0
        v2ant=v2
    else:
        ea=(((v2)-(v2ant))/(v2))*(100)
        v2ant=v2
        
        ev=v1-v2 
        erpv=(ev/v2)*100
        
    if(ea>es or ea==0):
        print "|   %-3d |   %-10f  |   %-10f  |   %-10f  |   %-12f|   %-10f  |"%(t,v1,v2,ev,erpv,ea)
        print ("-----------------------------------------------------------------------------------------")
    t+=i
    cont=ea 
```
